```python
"""
Data Mining & Warehousing - Practical Codes (1–6)
Complete implementations ready to run in Jupyter / Colab
"""

# ============================================================================
# PRACTICAL 1: DATA CLEANING & INTEGRATION
# ============================================================================

"""
Kaggle Dataset (Retail Alternative):
    Supermarket Sales
    https://www.kaggle.com/datasets/ponomarlili/sheet1?resource=download

Steps:
1. Download the dataset from Kaggle.
2. Save file as 'supermarket_sales.csv'.
3. Place it in the same folder as this script.

This practical demonstrates:
- Handling missing values
- Removing duplicates
- Renaming columns
- Splitting dataset into two sources (Village A & B)
- Merging them into integrated dataset
"""

import pandas as pd
import numpy as np

# Load Kaggle dataset
df = pd.read_csv("supermarket_sales.csv")

# Select and rename columns
df = df[['Branch', 'Product line', 'Quantity', 'Unit price']].copy()
df = df.rename(columns={
    'Branch': 'Village',
    'Product line': 'Product',
    'Unit price': 'Price'
})

# Split into Village A and Village B
village_a = df[df['Village'] == 'A'].copy()
village_b = df[df['Village'] == 'B'].copy()

print("BEFORE CLEANING:")
print(village_a.head(), "\n")
print(village_b.head(), "\n")

# Handle missing values
for col in ['Quantity', 'Price']:
    village_a[col] = village_a[col].fillna(village_a[col].mean())
    village_b[col] = village_b[col].fillna(village_b[col].mean())

# Remove duplicates
village_a = village_a.drop_duplicates()
village_b = village_b.drop_duplicates()

# Merge both villages
merged = pd.concat([village_a, village_b], ignore_index=True)

print("AFTER CLEANING & MERGING:")
print(merged.head())

# Save output
merged.to_csv("cleaned_supermarket_villages_A_B.csv", index=False)
print("\n✓ Saved cleaned dataset as cleaned_supermarket_villages_A_B.csv")
```

```python
"""
Data Mining & Warehousing - Practical Codes (1–6)
Complete implementations ready to run in Jupyter / Colab
"""

# ============================================================================
# PRACTICAL 2: DESCRIPTIVE STATISTICS
# ============================================================================

"""
This practical calculates:
- Mean
- Median
- Mode
- Variance
- Standard Deviation
- Quartiles (Q1, Q2, Q3)
- Range

Kaggle Dataset (Retail Alternative):
    Supermarket Sales
    https://www.kaggle.com/datasets/ponomarlili/sheet1?resource=download
"""

import pandas as pd

# Sample numeric column
data = pd.read_csv("supermarket_sales.csv")
print(data)

prices = data['Total']

print("\n" + "="*60)
print("PRACTICAL 2: DESCRIPTIVE STATISTICS")
print("="*60)

print(f"Mean:              {prices.mean():.2f}")
print(f"Median:            {prices.median():.2f}")
print(f"Mode:              {prices.mode()[0]:.2f}")
print(f"Variance:          {prices.var():.2f}")
print(f"Std Deviation:     {prices.std():.2f}")
print(f"Q1 (25%):          {prices.quantile(0.25):.2f}")
print(f"Q2 (50%):          {prices.quantile(0.50):.2f}")
print(f"Q3 (75%):          {prices.quantile(0.75):.2f}")
print(f"Range:             {prices.max() - prices.min():.2f}")

print("="*60)
```

```python
"""
Data Mining & Warehousing - Practical 3
DATA TRANSFORMATION
Includes:
- Normalization (Min-Max)
- Standardization (Z-Score)
- Equal-width binning (pandas + KBinsDiscretizer)
- Equal-frequency binning (pandas + KBinsDiscretizer)
"""

import pandas as pd
from sklearn.preprocessing import MinMaxScaler, StandardScaler, KBinsDiscretizer

# ---------------------------------------------------------------------------
# DATASET
# ---------------------------------------------------------------------------
data = pd.DataFrame({
    'Age':    [15, 22, 25, 29, 45, 58, 65],
    'Salary': [25000, 45000, 55000, 60000, 80000, 120000, 150000]
})

print("\n" + "="*60)
print("PRACTICAL 3: DATA TRANSFORMATION")
print("="*60)
print("ORIGINAL DATA:\n", data)

# ---------------------------------------------------------------------------
# 1. NORMALIZATION (Min-Max Scaling)
# ---------------------------------------------------------------------------
normalizer = MinMaxScaler()
normalized = data.copy()
normalized[['Age', 'Salary']] = normalizer.fit_transform(
    data[['Age', 'Salary']]
)

print("\n1. NORMALIZED DATA:\n", normalized)

# ---------------------------------------------------------------------------
# 2. STANDARDIZATION (Z-Score)
# ---------------------------------------------------------------------------
standardizer = StandardScaler()
standardized = data.copy()
standardized[['Age', 'Salary']] = standardizer.fit_transform(
    data[['Age', 'Salary']]
)

print("\n2. STANDARDIZED DATA:\n", standardized)

# ---------------------------------------------------------------------------
# 3. EQUAL-WIDTH BINNING (pandas)
# ---------------------------------------------------------------------------
binned = data.copy()
binned['Age_Bin'] = pd.cut(
    data['Age'],
    bins=3,                       # 3 bins (equal width)
    labels=['Young', 'Adult', 'Senior']
)

print("\n3. EQUAL-WIDTH BINNING (Age - pandas):\n", binned)

# ---------------------------------------------------------------------------
# 4. EQUAL-FREQUENCY BINNING (pandas)
# ---------------------------------------------------------------------------
freq_binned = data.copy()
freq_binned['Salary_Bin'] = pd.qcut(
    data['Salary'],
    q=3,                          # 3 quantile bins
    labels=['Low', 'Medium', 'High']
)

print("\n4. EQUAL-FREQUENCY BINNING (Salary - pandas):\n", freq_binned)

# ---------------------------------------------------------------------------
# 5. EQUAL-WIDTH BINNING (SKLEARN KBinsDiscretizer)
# ---------------------------------------------------------------------------
width_disc = KBinsDiscretizer(
    n_bins=3,                     # 3 bins
    encode='ordinal',             # output integers 0,1,2
    strategy='uniform'            # equal-width
)

width_binned = data.copy()
width_binned['Age_Bin'] = width_disc.fit_transform(
    data[['Age']]
).astype(int)

print("\n5. EQUAL-WIDTH BINNING (Age - KBinsDiscretizer):\n", width_binned)

# ---------------------------------------------------------------------------
# 6. EQUAL-FREQUENCY BINNING (SKLEARN KBinsDiscretizer)
# ---------------------------------------------------------------------------
freq_disc = KBinsDiscretizer(
    n_bins=3,
    encode='ordinal',
    strategy='quantile'           # equal-frequency
)

freq_binned2 = data.copy()
freq_binned2['Salary_Bin'] = freq_disc.fit_transform(
    data[['Salary']]
).astype(int)

print("\n6. EQUAL-FREQUENCY BINNING (Salary - KBinsDiscretizer):\n", freq_binned2)
```

```python

"""
Data Mining & Warehousing - Practical 4
ASSOCIATION RULE MINING (APRIORI)
Dataset: Iris Dataset
"""

import pandas as pd
from sklearn.datasets import load_iris
from mlxtend.frequent_patterns import apriori, association_rules

# ---------------------------------------------------------------------------
# 1. LOAD DATASET
# ---------------------------------------------------------------------------
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]

print("\n" + "="*60)
print("PRACTICAL 4: ASSOCIATION RULE MINING (APRIORI)")
print("="*60)
print("\nOriginal Iris Dataset (First 5 Rows):\n")
print(df.head())

# ---------------------------------------------------------------------------
# 2. DISCRETIZATION (Low / Medium / High)
# ---------------------------------------------------------------------------
df_cat = df.copy()

for col in iris.feature_names:
    df_cat[col] = pd.qcut(df[col], q=3, labels=['Low', 'Medium', 'High'], duplicates='drop')

print("\nDiscretized Dataset (First 5 Rows):\n")
print(df_cat.head())

# ---------------------------------------------------------------------------
# 3. ONE-HOT ENCODING
# ---------------------------------------------------------------------------
df_encoded = pd.get_dummies(df_cat)

print("\nOne-Hot Encoded Data (Columns):\n")
print(list(df_encoded.columns))

# ---------------------------------------------------------------------------
# 4. APRIORI ALGORITHM
# ---------------------------------------------------------------------------
frequent_itemsets = apriori(df_encoded, min_support=0.2, use_colnames=True)

print("\n" + "="*60)
print("FREQUENT ITEMSETS (min_support = 0.2)")
print("="*60)
print("\nTotal Frequent Itemsets Found:", len(frequent_itemsets))
print("\nTop Frequent Itemsets:\n")
print(frequent_itemsets.sort_values(by="support", ascending=False).head(10))

# ---------------------------------------------------------------------------
# 5. ASSOCIATION RULES
# ---------------------------------------------------------------------------
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)

print("\n" + "="*60)
print("ASSOCIATION RULES (confidence ≥ 0.5)")
print("="*60)

if len(rules) > 0:
    rules_display = rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].copy()
    rules_display['antecedents'] = rules_display['antecedents'].apply(lambda x: ', '.join(list(x)))
    rules_display['consequents'] = rules_display['consequents'].apply(lambda x: ', '.join(list(x)))

    print("\nNumber of Rules Generated:", len(rules_display))
    print("\nTop 10 Rules:\n")
    print(rules_display.head(10))
else:
    print("\nNo rules found at the given threshold.")
```

```python
"""
Data Mining & Warehousing - Practical Codes (1–6)
Complete implementations ready to run in Jupyter / Colab
"""

# ============================================================================
# PRACTICAL 5: CLASSIFICATION (Decision Tree vs Naive Bayes)
# ============================================================================

"""
Dataset: Iris Species

Compares:
- Decision Tree Classifier
- Gaussian Naive Bayes
"""

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print("\n" + "="*60)
print("PRACTICAL 5: CLASSIFICATION (Decision Tree & Naive Bayes)")
print("="*60)

# ---- Decision Tree ----
dt = DecisionTreeClassifier(random_state=42)
dt.fit(X_train, y_train)
pred_dt = dt.predict(X_test)

print("\nDecision Tree Accuracy:", accuracy_score(y_test, pred_dt))
print(confusion_matrix(y_test, pred_dt))
print(classification_report(y_test, pred_dt, target_names=iris.target_names))

# ---- Naive Bayes ----
nb = GaussianNB()
nb.fit(X_train, y_train)
pred_nb = nb.predict(X_test)

print("\nNaive Bayes Accuracy:", accuracy_score(y_test, pred_nb))
print(confusion_matrix(y_test, pred_nb))
print(classification_report(y_test, pred_nb, target_names=iris.target_names))

# ---- Comparison ----
print("\nModel Comparison:")
print(f"Decision Tree: {accuracy_score(y_test, pred_dt):.4f}")
print(f"Naive Bayes:   {accuracy_score(y_test, pred_nb):.4f}")
```

```python
"""
Data Mining & Warehousing - Practical Codes (1–6)
Complete implementations ready to run in Jupyter / Colab
"""

# ============================================================================
# PRACTICAL 6: CLUSTERING (K-Means & DBSCAN)
# ============================================================================

"""
Clustering using:
- K-Means
- DBSCAN

Dataset: Synthetic clusters generated using make_blobs()
"""

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
from sklearn.preprocessing import StandardScaler

# Generate synthetic data
X, y_true = make_blobs(
    n_samples=300,
    centers=4,
    cluster_std=0.6,
    random_state=42
)

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\n" + "="*60)
print("PRACTICAL 6: CLUSTERING (K-Means & DBSCAN)")
print("="*60)

# ---- K-MEANS ----
print("\nK-Means Clustering:")
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)
labels_kmeans = kmeans.fit_predict(X_scaled)

unique, counts = np.unique(labels_kmeans, return_counts=True)
for cluster, count in zip(unique, counts):
    print(f"Cluster {cluster}: {count} points")

# ---- DBSCAN ----
print("\nDBSCAN Clustering:")
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X_scaled)

n_clusters = len(set(labels_dbscan)) - (1 if -1 in labels_dbscan else 0)
n_noise = list(labels_dbscan).count(-1)

print(f"Clusters found (excluding noise): {n_clusters}")
print(f"Noise points: {n_noise}")

# ---- Plotting ----
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# K-Means plot
ax1.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels_kmeans, cmap='viridis', s=50)
ax1.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            c='red', marker='*', s=200, edgecolors='black')
ax1.set_title("K-Means Clustering")

# DBSCAN plot
colors = plt.cm.Spectral(np.linspace(0, 1, len(set(labels_dbscan))))
for label, color in zip(sorted(set(labels_dbscan)), colors):
    mask = (labels_dbscan == label)
    ax2.scatter(X_scaled[mask, 0], X_scaled[mask, 1], c=[color], s=50)
ax2.set_title("DBSCAN Clustering")

plt.tight_layout()
plt.show()
```
