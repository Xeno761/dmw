```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# 1. Load the Iris Dataset
iris_data = load_iris()

# 2. Create a Pandas DataFrame for better manipulation
# We combine the data (features) and target (species) into one table
df = pd.DataFrame(data=iris_data.data, columns=iris_data.feature_names)
df['species'] = iris_data.target  # Add target column

# Map the numerical target (0,1,2) to actual names for clarity
df['species'] = df['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# 3. Data Exploration
print("--- First 5 Rows ---")
print(df.head())  # Display first 5 rows

print("\n--- Dataset Description ---")
print(df.describe())  # Statistical summary (mean, std, min, max)

print("\n--- Class Distribution ---")
print(df['species'].value_counts())  # Count samples per class

# 4. Visualization
# Pairplot shows the relationship between every pair of features
print("\nGenerating Pairplot...")
sns.pairplot(df, hue='species', markers=["o", "s", "D"])
plt.title("Iris Dataset Visualization")
plt.show()
```
