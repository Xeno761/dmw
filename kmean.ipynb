```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate Synthetic Unlabeled Data
# n_samples=300, centers=4 (so we expect 4 clusters), random_state=42 for reproducibility
X, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=42)

# 2. Implement K-Means Clustering
# We choose K=4 because we generated 4 centers (in real life, use Elbow Method to find K)
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# 3. Fit the model
kmeans.fit(X)

# Get cluster labels and centroids
y_kmeans = kmeans.predict(X)
centroids = kmeans.cluster_centers_

# 4. Visualize Results
plt.figure(figsize=(8, 6))

# Plot data points colored by cluster
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis', label='Data Points')

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.7, marker='X', label='Centroids')

plt.title("K-Means Clustering Result")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
